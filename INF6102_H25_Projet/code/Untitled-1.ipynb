```python
import matplotlib.pyplot as plt

instances = ['A', 'B', 'C', 'D', 'E', 'Complet']
nb_pieces = [4*4, 7*7, 8*8, 9*9, 10*10, 16*16]

heuristic = [0, 11, 19, 26, 31, 81]
local_search = [0, 5, 17, 19, 21, 73]
ILS = [0, 0, 7, 10, 14, 54]
LNS = [0, 2, 9, 19, 20, 66]

percent_heur = [(nb_pieces[i]-heuristic[i])/nb_pieces[i] for i in range(len(instances))]
percent_ls = [(nb_pieces[i]-local_search[i])/nb_pieces[i] for i in range(len(instances))]
percent_ILS = [(nb_pieces[i]-ILS[i])/nb_pieces[i] for i in range(len(instances))]
percent_LNS = [(nb_pieces[i]-LNS[i])/nb_pieces[i] for i in range(len(instances))]

plt.plot(instances, percent_heur, label='heuristic')
plt.plot(instances, percent_ls, label='local search')
plt.plot(instances, percent_ILS, label='ILS')
plt.plot(instances, percent_LNS, label='LNS')

plt.legend()
plt.xlabel('instance')
plt.ylabel('pourcentage de complétion')
plt.title('Pourcentage de complétion obtenu par les différentes méthodes')
```
